```python
ncols = 4
nrows = int(np.ceil(len(smooth.index) / ncols))

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows*3))


def plot_axes(i,state):
    ax = axes.flat[i]
    ax.set_title(state)
    ax.set_ylim(0.0, 0.1)
    
    ax.plot(smooth.loc[:,'4/15/20':'6/3/20'].pct_change(axis=1).loc[state])    


for i, state in enumerate(smooth.index):
    plot_axes(i,state)

    
fig.tight_layout()
fig.set_facecolor('w')   
```
